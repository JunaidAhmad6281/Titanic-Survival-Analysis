# Import Libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

# Load Dataset

```python
df =pd.read_csv("titanic.csv")
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Finding % of null values in Age column 
(df["Age"].isna().sum()/len(df))*100
```

```python
df.duplicated().sum()
```

```python
# Handle missing values whit median of the age column 
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
# Filling the Embarked column with the value which is present most 
df["Embarked"].fillna("S", inplace = True)
```

```python
# we can drop Cabin column because it have lots of null values 
df.drop("Cabin",axis=1, inplace= True)
```

```python
# Now rechecking our data change or not 
df.isnull().sum()
```

```python
# Overall Survived rate
Survived_rate = df["Survived"].mean()*100
Survived_rate
```

```python
# Survived Count
df["Survived"].value_counts()
```

```python
# Survival Bar Chart 
df["Survived"].value_counts().plot(kind = "bar")
plt.title("Survival Count")
plt.xlabel("Survived (0 = NO, 1 = YES)")
plt.ylabel("Number Of Passengers")
plt.show()
```

```python
df["Age"].describe()
```

```python
plt.figure(figsize=(8, 5))
df["Age"].hist(bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(df["Age"].dropna())
plt.title("Age Box Plot")
plt.show()
```

```python
# Survival by Gender 
gender_survival = df.groupby("Gender")["Survived"].mean()*100
gender_survival
```

```python
plt.figure(figsize=(12,5))
ax = sns.countplot(data=df,x = 'Gender',hue='Survived')
plt.title("Survival Count By Gender (Male vs Female)")
plt.xlabel("Gender")
plt.ylabel("Number of Passenger")
plt.legend(title='Survived',labels=['Did not Survived (0)','Survived (1)'])
for container in ax.containers:
    ax.bar_label(container, fontsize=10)
plt.show
```

# Complete summary table

```python
gender_summary = df.groupby("Gender")["Survived"].agg(
    Total="count",
    Survived="sum",
    Survival_Rate=lambda x: round(x.mean() * 100, 2))
gender_summary
```
